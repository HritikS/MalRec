```python
from fastai import *
from fastai.tabular import * 
```

```python
path = '/home/hritik/Documents/BirdVision'
```

```python
df = pd.read_csv('data.csv', sep = '|')
df = df.drop(['Name', 'md5'], axis=1)
df.head()
```

```python
procs = [Normalize]
```

```python
valid_idx = range(len(df) - 27609, len(df))
```

```python
dep_var = 'legitimate'
```

```python
data = TabularDataBunch.from_df(path, df, dep_var, valid_idx=valid_idx, procs=procs)
data
```

```python
learn = tabular_learner(data, layers=[200, 100], metrics=accuracy)
learn
```

```python
learn.summary()
```

```python
learn.fit_one_cycle(6)
```
